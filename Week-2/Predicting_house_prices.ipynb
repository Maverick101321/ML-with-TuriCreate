# Simple regression model that predicts price from square feet

## Splitting the dataset into training and test data

```python
#Splitting the dataset into training_data and testing_data, keeping 80% for training and rest for testing
training_set, test_set = sales.random_split(.8,seed=0)
```

## Training a simple regression model

```python
# sqft_regression_model is the name of the model
# inside the bracket firstly we put the data to be used as input
# target defines the variable that we're trying to predict
# features defines what part(column) of the training data to be used as input
# in case we dont give anything in features then it will automatically use all columns of the data
sqft_regression_model = turicreate.linear_regression.create(training_set, target='price', features=['sqft_living'])
```

# Evaluating the quality of our model

```python
#printing the average price in the test dataset
print (test_set['price'].mean())
#we used print fn here cause it usually gives out the format in a proper way that is undersytandable
```

```python
#same cmd without print fn
test_set['price'].mean()
```

```python
#this gives out the 
#1. max error tells the maximum error between the test_data value and the predicted value
#2. root mean square error between the test data value and the predicted value
print (sqft_regression_model.evaluate(test_set))
#as we can see below that both the max error and the rmse are very high for the model
```

# Exploring the model a little further

```python
sqft_regression_model.coefficients
#the above line of code returns the coefficient of the linear regression model
#in other word the relationship between the features and the target variable
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
#the above line is to show all the plotting in this notebook
```

```python
plt.plot(test_set['sqft_living'], test_set['price'], '.', 
         test_set['sqft_living'], sqft_regression_model.predict(test_set), '-')
```
